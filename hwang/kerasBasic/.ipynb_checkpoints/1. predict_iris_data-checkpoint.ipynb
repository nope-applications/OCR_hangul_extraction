```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.utils import np_utils
import keras.backend.tensorflow_backend as K
import tensorflow as tf
import matplotlib.pyplot as plt
import os
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
session = tf.Session(config=config)
```

```python
data = pd.read_csv('./dataset/iris.csv', header=None)
print(data.head())
```

```python
X = data.iloc[:, 0:4]
y = data.iloc[:, 4]

print(set(y))
```

```python
#원핫 인코딩으로 바꿔준다

e = LabelEncoder()
e.fit(y)
y = e.transform(y)
y = np_utils.to_categorical(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(X_train.shape)
print(y_train.shape)
```

```python
with K.tf_ops.device('/device:GPU:0'):
    model = Sequential()
    model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(y_train.shape[1], activation='softmax'))

    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    model_dir = './model/'
    if not os.path.exists(model_dir):
        os.mkdir(model_dir)
    
    modelName = model_dir + 'iris.model'
    checkpointer = ModelCheckpoint(filepath=modelName, monitor='val_loss', verbose = 1, save_best_only=True)
    
    
    #학습의 자동중단! 5번동안 나아지지 않으면 stop!
    early_stop = EarlyStopping(monitor='val_loss', patience=5)
    
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=200, batch_size=10, callbacks=[checkpointer, early_stop])
```

```python
print("정확도 : %.2f " %(model.evaluate(X_test, y_test)[1]))
```

```python
y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len, y_vloss, marker='.', c='red', label='val_set_loss')
plt.plot(x_len, y_loss, marker='.', c='blue', label='train_set_oss')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()
plt.show()
```
